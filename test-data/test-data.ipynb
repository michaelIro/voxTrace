```python
import h5py as h5p
from matplotlib import pyplot as plt
import numpy as np
import xraylib as xrl
import os
```

```python
#filename = "/media/miro/Data-1TB/nist-1107-simulation/nist-1107-pos-(15.000000).h5"

# get number out of filename /media/miro/Data-1TB/nist-1107-simulation/nist-1107-pos-(NUMBER).h5
def get_file_number(filename):
    number = filename[57:-3].strip('()')
    number = int(float(number))
    return number

# find files in directory
directory = "/media/miro/Data-1TB/nist-1107-simulation"
files = [os.path.join(directory, file) for file in os.listdir(directory)]
files.sort(key=get_file_number)
#print(files)

n_files = len(files)
energies = []
respawns = []

i=0
# make list of h5p.File(file, 'r')['my_data'][20].sum() for each file
for file in files:
    #print(h5p.File(file, 'r')['my_data'][20].sum())
    respawns.append(h5p.File(file, 'r')['my_data'][20].sum())
    energies.append(h5p.File(file, 'r')['my_data'][10] / 50677300.0)
    i+=1
```

```python
for i in range(n_files):
    energies[i] /= respawns[i]
```

```python
# create list with int entries from 0 to n_files-1 
myrange = [i for i in range(n_files)]

##plot respawns over myrange
plt.ylabel("Respawns")
plt.xlabel("File #")
plt.plot(myrange, respawns, 'ro')
plt.show()
```
